```python
student = {
    'name': 'Muhammad Zoraiz Khan',
    'age': 20,
    'roll_number': 'SP23-BSE-105',
    'grade': 'B'
}


print("STEP 1: Print all keys and values")


print("Keys:", student.keys())


print("Values:", student.values())


print("\nKey-Value Pairs:")
for key, value in student.items():
    print(f"{key}: {value}")

print("\n\n")



print("STEP 2: Update the grade")

print(f"Old grade: {student['grade']}")
student['grade'] = 'A'
print(f"New grade: {student['grade']}")

print("\nUpdated dictionary:")
print(student)
print("\n\n")



print("STEP 3: Add email")


student['email'] = 'zoraizkhan@gmail.com'
print("Added email!")

print("\nDictionary after adding email:")
print(student)

print("\n\n")

print("STEP 4: Delete roll_number")


print(f"Deleting: roll_number = {student['roll_number']}")
student.pop('roll_number')

print("\nFinal dictionary:")
print(student)
```
